```python
for signal in SIGNALS:
    signal_name_parts = signal.split('_')
    true_mtop = int(signal_name_parts[-2])
    true_mchi = int(signal_name_parts[-1])

    histograms = a.results[signal].getScaledHistograms(s, None)
    obj = PlotObject.fromHist(histograms["m3_top_3_no_lead_b_pt_diff"])

    fig, ax = plt.subplots()
    drawAs1DHist(ax, obj, fill=False)
    
    ax.set_title(f"pt difference between chargino and other b\nmtop={true_mtop}, mchi={true_mchi}, key=m3_top_3...")
```
